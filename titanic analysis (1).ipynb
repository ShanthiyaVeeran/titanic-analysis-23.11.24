#DATA VISUALIZATION

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
x=np.array(['10','20','30','40','50'])
y=np.array(['100','45','90','86','21'])
```

```python
plt.title('overall Graph')
plt.xlabel('items')
plt.ylabel('available counts')
plt.plot(x,y,color='y')
plt.show()
```

```python
plt.title('overall Graph')
plt.xlabel('items')
plt.ylabel('available counts')
plt.plot(x,y,color='y',marker='o')
plt.show()
```
